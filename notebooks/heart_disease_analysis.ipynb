## Visualization


Age Distribution

```python
plt.hist(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Sex Distribution

```python
df["sex"].value_counts()
```

```python
sns.countplot(x="sex", data=df)

plt.title("Distribution of Patients by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")

plt.show()
```

## Heart Disease by Sex

```python
df["target"].value_counts()
```

```python
df["heart_disease"]=(df["target"]> 0).astype(int)
```

```python
df["heart_disease"].value_counts()
```

```python
pd.crosstab(df["sex"], df["heart_disease"])
```

Visualization

```python


sns.countplot(
    x="sex",
    hue="heart_disease",
    data=df
)

plt.title("Heart Disease by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Number of Patients")

plt.show()
```

```python
columns = [
    "age","sex","cp","trestbps","chol",
    "fbs","restecg","thalach","exang",
    "oldpeak","slope","ca","thal","target"
]

df = pd.read_csv(
    "../data/processed.cleveland.data",
    names=columns
)

df.head()
```

## Visualization


Age Distribution

```python
plt.hist(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Sex Distribution

```python
df["sex"].value_counts()
```

```python
sns.countplot(x="sex", data=df)

plt.title("Distribution of Patients by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")

plt.show()
```

```python
df["sex"].value_counts(normalize=True)*100
```

## Heart Disease by Sex

```python
df["target"].value_counts()
```

```python
df["heart_disease"]=(df["target"]> 0).astype(int)
```

```python
df["heart_disease"].value_counts()
```

```python
pd.crosstab(df["sex"], df["heart_disease"])
```

Visualization

```python


sns.countplot(
    x="sex",
    hue="heart_disease",
    data=df
)

plt.title("Heart Disease by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Number of Patients")

plt.savefig("../figures/heart_disease_by_sex.png")

plt.show()
```

## Heart Disease by Age 

```python
df.groupby("heart_disease")["age"].describe()
```

```python
sns.boxplot(
    x="heart_disease",
    y="age",
    data=df
)


plt.title("Age Distribution by Heart Disease Status")
plt.xlabel("Heart Disease")
plt.ylabel("Age")

plt.savefig("../figures/heart_disease_by_age.png")

plt.show()
```

```python
from scipy.stats import ttest_ind

healthy = df[df["heart_disease"] == 0]["age"]
disease = df[df["heart_disease"] == 1]["age"]

ttest_ind(healthy, disease)
```
